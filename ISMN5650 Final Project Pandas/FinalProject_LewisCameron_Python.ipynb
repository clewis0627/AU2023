```python
import pandas as pd

math = pd.read_csv("SchoolMath.csv")
```

```python
# Allows for all columns to be visible for additional data analysis
pd.set_option('display.max_columns', None)

math.head()
```

```python
math.tail()
```

```python
math.info()
```

```python
math.describe()
```

```python
# Drop NaN values
math = math.dropna()
math.head()
```

```python
# Drop dupes where the below value set trio is repeated (ex. SchoolA and 3 and 2014)
math = math.drop_duplicates(subset=["School Name", "Grade", "Year"])
math
```

```python
pattern = r'^P\.S\. \d{3} '

# Replace values in the 'school_name' column
math['School Name'] = math['School Name'].str.replace(pattern, '')
roberto = math[math["School Name"] == "Roberto Clemente"]

roberto
```

```python
# Drop dupes where the below value set trio is repeated (ex. SchoolA and 3 and 2014)
roberto = roberto.drop_duplicates(subset=["School Name", "Grade", "Year"])

roberto = roberto.sort_values(by=['Grade', 'Year'])

roberto
```

```python
# Convert object datatypes into numeric datatypes for analysis

roberto[roberto["Mean Scale Score"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 1 #"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 1 %"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 2 #"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 2 %"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 3 #"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 3 %"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 4 #"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 4 %"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 3+4 #"].apply(lambda x: str(x).isdigit())]
roberto[roberto["Level 3+4 %"].apply(lambda x: str(x).isdigit())]


roberto["Mean Scale Score"] = pd.to_numeric(math["Mean Scale Score"], errors="coerce")
roberto["Level 1 #"] = pd.to_numeric(roberto["Level 3 #"], errors="coerce")
roberto["Level 1 %"] = pd.to_numeric(roberto["Level 3 %"], errors="coerce")
roberto["Level 2 #"] = pd.to_numeric(roberto["Level 3 #"], errors="coerce")
roberto["Level 2 %"] = pd.to_numeric(roberto["Level 3 %"], errors="coerce")
roberto["Level 3 #"] = pd.to_numeric(roberto["Level 3 #"], errors="coerce")
roberto["Level 3 %"] = pd.to_numeric(roberto["Level 3 %"], errors="coerce")
roberto["Level 4 #"] = pd.to_numeric(roberto["Level 4 #"], errors="coerce")
roberto["Level 4 %"] = pd.to_numeric(roberto["Level 3 %"], errors="coerce")
roberto["Level 3+4 #"] = pd.to_numeric(roberto["Level 3 #"], errors="coerce")
roberto["Level 3+4 %"] = pd.to_numeric(roberto["Level 3 %"], errors="coerce")

# Drop NaN values from converting to numeric
roberto = roberto.dropna()

# Display results to confirm changes
roberto
```

```python
# Get info to confirm changes were successful
roberto.info()
```

```python
# Import neccessary libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Looks specifically at 4th grade Mean Scale Score data over the 6 years of data available

grade4 = roberto[roberto["Grade"] == "4"]

sns.lmplot(x="Year", y="Mean Scale Score", data=grade4)
```

```python
# Plots The Mean Scale Score by Year per Grade
f, axs = plt.subplots(1, 1, figsize=(8, 8))
sns.lineplot(data=roberto, x="Year", y="Mean Scale Score", hue="Grade", ax=axs)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
